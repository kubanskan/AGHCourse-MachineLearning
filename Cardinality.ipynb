# Zadanie 5

```python
df['CabinReduced'] = df['cabin'].astype(str).str[0]
```

```python
df[['cabin', 'CabinReduced']].head(20)
```

# Zadanie 6

```python
print(f'Liczba etykiet zmiennej  CabinReduced: {len(df["CabinReduced"].unique())}')
```

```python
reduction =  ( len(df["cabin"].unique()) - len(df["CabinReduced"].unique()) ) / len(df["cabin"].unique()) * 100
print(f'Kardynalność została zredukowana o {reduction:.2f}%')
```
